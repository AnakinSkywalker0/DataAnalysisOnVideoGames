# Video Games Sales Dataset Overview

## Basic Information
- **Dataset Name:** gamesdata.csv  
- **Total Rows:** 16,719  
- **Total Columns:** 16  

This dataset contains information about video games, their platforms, sales performance across different regions, and review scores from critics and users. It can be used to analyze sales patterns, rating influences, and genre trends across years.

---

## Columns Description

| Column Name | Description | Data Type | Example |
|--------------|-------------|------------|----------|
| **Name** | Title of the video game | Object | "Wii Sports" |
| **Platform** | Gaming console or system | Object | "Wii" |
| **Year_of_Release** | Release year of the game | Float | 2006.0 |
| **Genre** | Type of game | Object | "Sports" |
| **Publisher** | Company that published the game | Object | "Nintendo" |
| **NA_Sales** | Sales in North America (in millions) | Float | 41.36 |
| **EU_Sales** | Sales in Europe (in millions) | Float | 28.96 |
| **JP_Sales** | Sales in Japan (in millions) | Float | 3.77 |
| **Other_Sales** | Sales in other regions (in millions) | Float | 8.45 |
| **Global_Sales** | Total worldwide sales (in millions) | Float | 82.53 |
| **Critic_Score** | Average critic rating (0–100) | Float | 76.0 |
| **Critic_Count** | Number of critic reviews | Float | 51.0 |
| **User_Score** | Average user rating (0–10) | Object | "8.0" |
| **User_Count** | Number of user ratings | Float | 322.0 |
| **Developer** | Company that developed the game | Object | "Nintendo" |
| **Rating** | ESRB content rating (E, M, T, etc.) | Object | "E" |

---

## Notes
- Some columns contain **missing values**, especially review-related ones (Critic_Score, User_Score, Developer, Rating).  
- The dataset is well-suited for analysis of:
  - Global and regional sales performance  
  - Relationship between scores and sales  
  - Platform and genre popularity trends over time
