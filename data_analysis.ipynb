# Video Games Sales Dataset Overview

## Basic Information
- **Dataset Name:** gamesdata.csv  
- **Total Rows:** 16,719  
- **Total Columns:** 16  

This dataset contains information about video games, their platforms, sales performance across different regions, and review scores from critics and users. It can be used to analyze sales patterns, rating influences, and genre trends across years.

---

## Columns Description

| Column Name | Description | Data Type | Example |
|--------------|-------------|------------|----------|
| **Name** | Title of the video game | Object | "Wii Sports" |
| **Platform** | Gaming console or system | Object | "Wii" |
| **Year_of_Release** | Release year of the game | Float | 2006.0 |
| **Genre** | Type of game | Object | "Sports" |
| **Publisher** | Company that published the game | Object | "Nintendo" |
| **NA_Sales** | Sales in North America (in millions) | Float | 41.36 |
| **EU_Sales** | Sales in Europe (in millions) | Float | 28.96 |
| **JP_Sales** | Sales in Japan (in millions) | Float | 3.77 |
| **Other_Sales** | Sales in other regions (in millions) | Float | 8.45 |
| **Global_Sales** | Total worldwide sales (in millions) | Float | 82.53 |
| **Critic_Score** | Average critic rating (0–100) | Float | 76.0 |
| **Critic_Count** | Number of critic reviews | Float | 51.0 |
| **User_Score** | Average user rating (0–10) | Object | "8.0" |
| **User_Count** | Number of user ratings | Float | 322.0 |
| **Developer** | Company that developed the game | Object | "Nintendo" |
| **Rating** | ESRB content rating (E, M, T, etc.) | Object | "E" |

---

## Notes
- Some columns contain **missing values**, especially review-related ones (Critic_Score, User_Score, Developer, Rating).  
- The dataset is well-suited for analysis of:
  - Global and regional sales performance  
  - Relationship between scores and sales  
  - Platform and genre popularity trends over time  

---



# Importing Data On Video Games

```python
import pandas as pd
```

```python

df = pd.read_csv("gamesdata.csv")
df.head(10)
```

# Basic Information About The Data


```python
df.describe()
```

```python
df.info()
```

# Checking Null Values 

```python
df.isnull().sum()
```

# Removing problematic rows Inside missing value category 

```python

df = df.dropna(subset=['Year_of_Release'])
df['Year_of_Release'] = df['Year_of_Release'].astype(int)
df = df.dropna(subset=['Name', 'Genre'])
df['Publisher'] = df['Publisher'].fillna('Unknown')
```

# 

## Checking Changes 

```python
df.isnull().sum()
df.info()
```

# Actual Data Analysis Start From Here 

```python
import matplotlib.pyplot as plt

sales_per_year = df.groupby('Year_of_Release')['Global_Sales'].sum()

plt.figure(figsize=(12,6))

plt.plot(
    sales_per_year.index,
    sales_per_year.values,
    color='#00b3ff',      # bright cyber blue line
    linewidth=3
)

# Title and labels with strong colors
plt.title(
    'Global Video Game Sales Over the Years',
    fontsize=20,
    color='#ff5733',      # bold orange-red
    pad=15
)

plt.xlabel(
    'Year of Release',
    fontsize=14,
    color='#4CAF50'       # green
)

plt.ylabel(
    'Global Sales (Millions)',
    fontsize=14,
    color='#4CAF50'
)

# Tick styling
plt.xticks(fontsize=12, color='white', rotation=45)
plt.yticks(fontsize=12, color='white')

# Grid styling
plt.grid(True, alpha=0.3, color='gray')

# Background & axes styling
plt.gca().set_facecolor('#1e1e1e')      # dark background for the plot area
plt.gcf().patch.set_facecolor('#111111') # dark background for full figure
plt.tight_layout()

plt.show()
```

```python
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(12,6))
genre_counts.plot(kind='bar', color='#00b3ff')
plt.title('Number of Games Released per Genre', fontsize=16, color='#ff5733')
plt.xlabel('Genre', fontsize=12, color='#4CAF50')
plt.ylabel('Count', fontsize=12, color='#4CAF50')
plt.xticks(rotation=45, fontsize=10, color='white')
plt.yticks(fontsize=10, color='white')
plt.grid(axis='y', alpha=0.3)
plt.gca().set_facecolor('#1e1e1e')
plt.gcf().patch.set_facecolor('#111111')
plt.tight_layout()
plt.show()
```

# Games Per Genre


```python
genre_year = df.groupby(['Year_of_Release', 'Genre']).size().unstack(fill_value=0)

plt.figure(figsize=(14,6))
genre_year.plot(kind='line', figsize=(14,6))
plt.title('Genre Release Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# SEPERATING THE TOP 5 GENRES 

```python
top5_genres = df['Genre'].value_counts().head(5).index
top5_genres

```

```python
df_top5 = df[df['Genre'].isin(top5_genres)]  # Filtered top 5 genres 
```

```python
genre_year_top5 = df_top5.groupby(['Year_of_Release', 'Genre']).size().unstack(fill_value=0)

plt.figure(figsize=(14,6))

for genre in genre_year_top5.columns:
    plt.plot(genre_year_top5.index, genre_year_top5[genre], label=genre)

plt.title('Top 5 Genres: Number of Games Released Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.grid(alpha=0.3)
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

genre_avg_sales = df.groupby('Genre')['Global_Sales'].mean().sort_values(ascending=False)

plt.figure(figsize=(14,7))
plt.bar(genre_avg_sales.index, genre_avg_sales.values, color='#00e676')

plt.title('Average Global Sales Per Game (Genre Profitability)', fontsize=20, color='#ffca28')
plt.xlabel('Genre', fontsize=14, color='#4dd0e1')
plt.ylabel('Avg Sales (Millions)', fontsize=14, color='#4dd0e1')

plt.xticks(rotation=45, fontsize=11, color='white')
plt.yticks(fontsize=11, color='white')
plt.grid(axis='y', alpha=0.3, color='gray')

plt.gca().set_facecolor('#1e1e1e')
plt.gcf().patch.set_facecolor('#121212')

plt.tight_layout()
plt.show()
```

# Global Sales Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))

sns.kdeplot(
    df['Global_Sales'],
    fill=True,
    linewidth=3,
    color='#4A90E2'
)

plt.title('Distribution of Global Video Game Sales', fontsize=22, color='#333')
plt.xlabel('Global Sales', fontsize=16, color='#444')
plt.ylabel('Density', fontsize=16, color='#444')

plt.xticks(fontsize=12, color='#222')
plt.yticks(fontsize=12, color='#222')

plt.grid(alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```
